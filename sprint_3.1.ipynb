```python
import pytesseract
from pytesseract import Output
import cv2
img = cv2.imread('c1.png')

d = pytesseract.image_to_data(img, output_type=Output.DICT)
n_boxes = len(d['level'])
bbox_list =[]
bboxtxt_list=[]
x_max=[]
y_max=[]
line_num=[]
for i in range(n_boxes):
    if(d['text'][i] != ""):
        (x, y, w, h) = (d['left'][i], d['top'][i], d['width'][i], d['height'][i])
        x_max.append(d['left'][i]+d['width'][i])
        y_max.append(d['top'][i]+d['height'][i])
        line_num.append(d['par_num'][i])
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
        bbox = (x, y, w, h)
        bbox_list.append(bbox)
        bboxtxt_list.append(d['text'][i])
        print(d['text'][i])
print(bbox_list)
print(x_max)
print(y_max)
print("line_num:" ,line_num)
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

filepath = "1.png"

img = Image.open(filepath).convert('L')
WIDTH, HEIGHT = img.size
#print("1:::",WIDTH,HEIGHT,":::1")
pix = img.load()
#print("1:::",pix,":::1")
data = np.asarray(img.getdata())
#print("1:::",data,":::1")
data = data.reshape((HEIGHT,WIDTH))
#print("1:::",data,":::1")
fig,ax = plt.subplots()
reduced_data = data.mean(axis=0)
print("1:::",reduced_data,":::1")
ax.plot(reduced_data)

plt.show()
plt.imshow(img)
```
