```python
import pytesseract
from pytesseract import Output
import cv2
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import glob
import pandas as pd
import time
```

```python
def grayscale_plot(filepath):
    otsu_mag(filepath)
    img = Image.open(filepath).convert('L')
    WIDTH, HEIGHT = img.size
    #print("1:::",WIDTH,HEIGHT,":::1")
    pix = img.load()
    #print("1:::",pix,":::1")
    data = np.asarray(img.getdata())
    #print("1:::",data,":::1")
    data = data.reshape((HEIGHT,WIDTH))
    #print("1:::",data,":::1")
    #fig,ax = plt.subplots()
    reduced_data = data.mean(axis=0)
    #print("1:::",reduced_data,":::1")
    #ax.plot(reduced_data)
    #plt.show()
    #plt.imshow(img)
    return reduced_data,WIDTH,HEIGHT

def c_sep_size1(reduced_data,WIDTH):
    p=[]
    c_sep=[]
    c_sep1=[]
    c_sep_size=[]
    for i in range(11,len(reduced_data)-10):
        if(reduced_data[i-3]==255. and reduced_data[i+3]==255.):
            p.append(i)
    for a in range(len(p)-1):
        if(p[a+1]-p[a]>30):
            c_sep.append(p[a])
    if(len(p)>0):
        c_sep.append(p[-1])

    for i in range(0,len(c_sep)-1):
        if(c_sep[i+1]-c_sep[i]!=1):
            c_sep1.append(c_sep[i])
    if(len(c_sep)>0):
        c_sep1.append(c_sep[-1])
    c_sep1.append(WIDTH)
    c_sep_size.append(c_sep1[0])
    for i in range(len(c_sep1)-1):
        c_sep_size.append(c_sep1[i+1]-c_sep1[i])
    #c_sep_size.append(c_sep[-1]-c_sep1[-1])
    return c_sep,c_sep1,c_sep_size
```

```python

def tess_ract(filepath):
    img = cv2.imread(filepath)
    d = pytesseract.image_to_data(img, output_type=Output.DICT)
    n_boxes = len(d['level'])
    bbox_list =[]
    bboxtxt_list=[]
    vall=[]
    for i in range(n_boxes):
        if(d['text'][i] != ""):
            (x, y) = (d['left'][i], d['top'][i])
            #cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
            #bbox = (x, y, w, h)
            #bbox_list.append(bbox)
            #bboxtxt_list.append(d['text'][i])
            #print(d['text'][i])
            vall.append([x,d['text'][i]])
    #cv2.imshow('img', img)
    #cv2.waitKey(0)
    #cv2.destroyAllWindows()
    return vall

def final_opp(c_sep1,vall):
    final_op={}
    q=0
    m=[]
    x=0
    for j in range(len(c_sep1)):
        for i in vall:
            if(i[0]<c_sep1[j] and i[0]>=x):
            
                m.append(i[1])
        x=c_sep1[j]
        if(len(m)!=0):
            final_op[q]=m
            m=[]
            q+=1
    return final_op
    
    
def all_file(root_dir,op):
    headers=[]
    img_name=[]
    time_ellapsed=[]
    num_columns=[]
    k=[]
    l=[]
    for filename in glob.iglob(root_dir + '**/*.jpg', recursive=True):
        
        val = tess_ract(filename)
        t0= time.time()
        reduced_data,WIDTH,HEIGHT=grayscale_plot(filename)
        c_sep,c_sep1,c_sep_size=c_sep_size1(reduced_data,WIDTH)
        final_op = final_opp(c_sep1,val)
        t1 = time.time() - t0
        k.append(filename)
        k.append(len(final_op))
        k.append(final_op)
        k.append(t1)
        #df = into_df(img_name,num_columns,headers,time_ellapsed)
        l.append(k)
        k=[]
        df = pd.DataFrame(l,columns =['img_name','num_columns','headers','time_ellapsed']) 
        to_csv(df,op)
    return df

# def into_df(img_name,num_columns,headers,time_ellapsed):
#     df = pd.DataFrame((list(zip(img_name,num_columns,headers,time_ellapsed)),columns =['img_name','num_columns','headers','time_ellapsed'])
#     return d


def otsu_mag(filepath):
    image1 = cv2.imread(filepath)
    img = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
    ret, thresh1 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)     
    #print(thresh1)
    if(thresh1.mean()<=30):
        thresh1 =255- thresh1  
    #cv2.imshow('Otsu Threshold', thresh1)     
    cv2.imwrite(filepath, thresh1)      

    if cv2.waitKey(0) & 0xff == 27:
        cv2.destroyAllWindows()     
        
def to_csv(df,op):
    df.to_csv(op,index=False)
    
```
